#### Question 6 - For the remaining rows which have no record (but the other columns *are* populated), determine an appropriate method for inputting the value for this column. Explain your reasoning, and show your work below (using pandas).

```python
"""
My method is to use some of the temperature data to determine if the No Record category is most likely Full, Partial, or No Shadow.
The median differences vary enough to between Full, Partial, and No Shadow, so I use these values and compare with each of the No Record rows.
Step 1: Group by No Shadow, Partial Shadow, Full Shadow
Step 2: find the median difference in temperature between March and Februrary Average Temperatures
Step 3: determine appropriate label for No Record:            
"""

#Group by No Shadow, Partial Shadow, Full Shadow
no_shadow = df.loc[(df['Punxsutawney Phil'] == 'No Shadow')]
p_shadow = df.loc[(df['Punxsutawney Phil'] == 'Partial Shadow')]
f_shadow = df.loc[(df['Punxsutawney Phil'] == 'Full Shadow')]

#find median difference between March and Februrary Average Temperatures
full_shadow_median_tempChange = (f_shadow['March Average Temperature (Pennsylvania)'] -f_shadow['February Average Temperature (Pennsylvania)']).median()
part_shadow_median_tempChange = (p_shadow['March Average Temperature (Pennsylvania)'] -p_shadow['February Average Temperature (Pennsylvania)']).median()
no_shadow_median_tempChange = (no_shadow['March Average Temperature (Pennsylvania)'] -no_shadow['February Average Temperature (Pennsylvania)']).median()

#use medians to determine if the No Record values are <= the median values
df.loc[(df['Punxsutawney Phil'] == 'No Record') &  
       ((df['March Average Temperature (Pennsylvania)'] -df['February Average Temperature (Pennsylvania)']).median() <= no_shadow_median_tempChange), "Punxsutawney Phil"] = 'No Shadow'
df.loc[(df['Punxsutawney Phil'] == 'No Record') &  
       ((df['March Average Temperature (Pennsylvania)'] -df['February Average Temperature (Pennsylvania)']).median() <= part_shadow_median_tempChange), "Punxsutawney Phil"] = 'Partial Shadow'
df.loc[(df['Punxsutawney Phil'] == 'No Record') &  
       ((df['March Average Temperature (Pennsylvania)'] -df['February Average Temperature (Pennsylvania)']).median() > part_shadow_median_tempChange), "Punxsutawney Phil"] = 'Full Shadow'

df
```
