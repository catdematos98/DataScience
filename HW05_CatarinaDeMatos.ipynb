```python
import pandas as pd
import numpy as np

##Visualization Libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

data_dir = 'C:\\Users\\CAT\\Desktop\\Python\\'
weather_file = data_dir + 'student_class.csv'
df = pd.read_csv(weather_file)
df.head()
```

```python
#1
major_groups = df.groupby('Majors')
midterm_median = major_groups['Midterm'].agg(np.median)
final_median = major_groups['Final'].agg(np.median)
finalAverage_median = major_groups['Final Average'].agg(np.median)

labels = ['BNFO', 'CS', 'IS', 'IT', 'MATH']
x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

plt.bar(x-width, midterm_median, width, label='Midterm')
plt.bar(x, final_median, width, label='Final')
plt.bar(x+width, finalAverage_median, width, label='Final Average')
plt.gcf().set_size_inches(12, 6)

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.ylabel('Major')
plt.xlabel('Median Score')
plt.title('Midterm, Final, and Final Average Median Score by Major')
plt.xticks(x, labels)
plt.legend(loc='upper right')
plt.show()
```

```python
#2
n, bins, patches = plt.hist(df['Numeric Bin'], 6)
plt.gcf().set_size_inches(12, 6)

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.ylabel('Frequency')
plt.xlabel('Grade')
plt.title('Number of students in each grade bin')
plt.show()
```

```python
#3
plt.plot(df['Final'], df['Final Average'], 'o', c="blue", label="Final")
plt.plot(df['Midterm'], df['Final Average'], 'o', c="red", label="Midterm")
plt.gcf().set_size_inches(12, 6)
plt.ylabel('Final Average')
plt.xlabel('Exam Score')
plt.title('Midterm and Final Exam Scores vs. Final Average')
plt.tight_layout()
plt.legend()
plt.show()
```
